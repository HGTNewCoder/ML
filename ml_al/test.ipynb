```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad(x):
    return 2*x + 5*np.cos(x)
def f(x):
    return x**2 + 5*np.sin(x)
def gd(epoch, lr, x0):
    x = x0 - lr*grad(x0)
    if epoch == 0:
        return x
    return gd(epoch-1, lr, x)

optimal_value = gd(1000, 0.01, 5)
print(optimal_value)
def plot():
    x = np.linspace(-6, 6)
    y = f(x)
    z = grad(x)
    plt.title("gradient descent test")
    plt.ylabel("f(x)")
    plt.xlabel("x")
    plt.plot(x, y)
    plt.plot(x, z)
    plt.scatter(optimal_value, f(optimal_value), c = 'red', marker='*')
    plt.show()
    
plot()
```
